```python
# importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd
from datetime import datetime, timedelta
import time
```

```python
start = time.time()

# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('facebook data since:2018-01-01 until:2018-01-02').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content])
    
# Creating a dataframe from the tweets list above
df_tweets = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text'])

end = time.time()
print(end - start)
```

```python
def twitter_scapper(keyword, since, until):
    tweets_list = []
    print(f"{keyword} since:{since} until:{until}")
    tweets = sntwitter.TwitterSearchScraper(f"{keyword} since:{since} until:{until}")
    for i,tweet in enumerate(tweets.get_items()):
        print(tweet.date)
        tweets_list.append([tweet.date, tweet.id, tweet.content])
    df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text'])
    df.to_csv(f"../data/tweets/{keyword}_{since}_{until}")
```

```python
def seg_date(keyword, since, until):
    d_since = pd.to_datetime(since, format='%Y-%m-%d')
    d_until = pd.to_datetime(until, format='%Y-%m-%d')
    num_days = (d_until-d_since).days
    while num_days > 15:
        start = time.time()
        print(f"{num_days} days remaining")
        until = d_since + timedelta(days=15)
        twitter_scapper(keyword, d_since.strftime('%Y-%m-%d'), until.strftime('%Y-%m-%d'))
        d_since = until
        num_days = (d_until-d_since).days
        end = time.time()
        print(end - start)
    start = time.time()
    print(f"{num_days} days remaining")
    twitter_scapper(keyword, d_since.strftime('%Y-%m-%d'), d_until.strftime('%Y-%m-%d'))
    d_since = until
    num_days = (d_until-d_since).days
    end = time.time()
    print(end - start)
```

```python
seg_date("kinder", "2018-01-01", "2018-02-01")
```
